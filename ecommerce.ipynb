```python
#dependecias
import pandas as pd
from pandas.api.types import is_numeric_dtype
```

```python
#configuraciones
pd.set_option('display.max_columns', None)
```

```python
#datos
reviews= pd.read_csv('olist_order_reviews_dataset.csv')
orders = pd.read_csv('olist_orders_dataset.csv')
customers= pd.read_csv('olist_customers_dataset.csv')
geolocation= pd.read_csv('olist_geolocation_dataset.csv')
items= pd.read_csv('olist_order_items_dataset.csv')
payments= pd.read_csv('olist_order_payments_dataset.csv')
products= pd.read_csv('olist_products_dataset.csv')
sellers= pd.read_csv('olist_sellers_dataset.csv')
translation= pd.read_csv('product_category_name_translation.csv')
```

```python
conjunto = {
    'reviews': 'olist_order_reviews_dataset.csv',
    'orders': 'olist_orders_dataset.csv',
    'customers': 'olist_customers_dataset.csv',
    'geolocation': 'olist_geolocation_dataset.csv',
    'items': 'olist_order_items_dataset.csv',
    'payments': 'olist_order_payments_dataset.csv',
    'products': 'olist_products_dataset.csv',
    'sellers': 'olist_sellers_dataset.csv',
    'translation': 'product_category_name_translation.csv'
}

for name, file in conjunto.items():
    print(f"Nombres de columnas de {name}:")
    df = pd.read_csv(file)
    print(', '.join(df.columns))
    print()
```

```python
full = pd.merge(reviews, orders, on='order_id', how='left')

#customers & geolocation
full = pd.merge(full, customers, on='customer_id', how='left')
full = pd.merge(full, geolocation, left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix', how='left')

#items & payments
full = pd.merge(full, items, on='order_id', how='left')
full = pd.merge(full, payments, on='order_id', how='left')

#products & sellers
full = pd.merge(full, products, on='product_id', how='left')
full = pd.merge(full, sellers, on='seller_id', how='left')

#translation
full = pd.merge(full, translation, on='product_category_name', how='left')
```

```python
def univariado_info(full):
    '''Calculo de informacion estadisticas y genericas de cada columna del dataframe'''

    df_info = pd.DataFrame(columns=['Cantidad', 'Tipo' , 'Missing', 'Unicos', 'Numeric'])
    for col in full:
        data_series = full[col]
        if is_numeric_dtype(data_series):
            df_info.loc[col] = [data_series.count(), data_series.dtype, data_series.isnull().sum(), data_series.nunique(), True]
        else:
            df_info.loc[col] = [data_series.count(), data_series.dtype, data_series.isnull().sum(), data_series.nunique(), False]

    numeric_columns = full.select_dtypes(include=['float64', 'int64'])
    df_describe = numeric_columns.describe().T[['mean', 'std', 'min', '25%', '50%', '75%', 'max']]
    df_stats = pd.DataFrame([numeric_columns.skew(), numeric_columns.kurtosis()], index=['sesgo', 'kurt']).T

    return pd.concat([df_info, pd.concat([df_describe, df_stats], axis=1)], axis=1).fillna('-')

full_stats = univariado_info(full)
full_stats
```
